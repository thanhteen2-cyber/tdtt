```python
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0: return False
    return True
```

```python
def is_perfect(n):
    if n <= 0: return False
    return sum(i for i in range(1, n) if n % i == 0) == n
```

```python
def factorial(n):
    res = 1
    for i in range(2, n + 1):
        res *= i
    return res
```

```python
def calculator(num1, num2, operator):
    if operator == '+': res = num1 + num2
    elif operator == '-': res = num1 - num2
    elif operator == '*': res = num1 * num2
    elif operator == '/': res = num1 / num2 if num2 != 0 else "Error"
    else: return "Invalid operator"
    
    return round(res, 2) if isinstance(res, (int, float)) else res
```
