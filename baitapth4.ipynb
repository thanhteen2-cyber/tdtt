```python
n = int(input())
max = 0
best = 0
for i in range(1, n + 1):
    xet = i
    dodai = 1
    while xet > 1:
        if xet % 2 == 0:
            xet //= 2
        else:
            xet = 3 * xet + 1
        dodai += 1
    if dodai > max:
        max = dodai
        best = i
print(best , max)
```
