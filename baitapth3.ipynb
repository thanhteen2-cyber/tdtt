```python
x = float(input())
if x == int(x):
    ceil_val = int(x)
elif x > 0:
    ceil_val = int(x) + 1
else:
    ceil_val = int(x)

if x == int(x):
    floor_val = int(x)
elif x > 0:
    floor_val = int(x)
else:
    floor_val = int(x) - 1

if x >= 0:
    if x - int(x) >= 0.5:
        round_val = int(x) + 1
    else:
        round_val = int(x)
else:
    if x - int(x) <= -0.5:
        round_val = int(x) - 1
    else:
        round_val = int(x)
print(ceil_val, floor_val, round_val)
```

```python
a, b, c, d = map(int, input().split())
if a == 0:
    if b == 0 and c == 0 and d == 0:
        result = "YES"
    else:
        result = "NO"
else:
    if b % a == 0:
        q = b // a
        if b * q == c and c * q == d:
            result = "YES"
        else:
            result = "NO"
    else:
        result = "NO"

print(result)
```
